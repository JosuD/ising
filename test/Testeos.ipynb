```python
import numpy as np
from matplotlib import pyplot as plt

data = np.genfromtxt('test.txt', delimiter = '\t')
bines = np.arange(64)
#rep = np.histogram(data, bins = 64)
plt.hist(data, bins = 64)
print(len(data), np.mean(data), np.std(data))
plt.title('Distribucion de los indices de pick_site, para L = 8')
plt.xlabel('Indice')
plt.ylabel('Frecuencia')
plt.text(40,50, 'mean = %.2f'%(np.mean(data)))
plt.text(40,25, 'std = %.2f'%(np.std(data)))
plt.savefig('test_pick_site.png',format = 'png', dpi = 200)

plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

energia = np.genfromtxt('energia.txt', delimiter = '\t')
magnet = np.genfromtxt('magnetizacion.txt', delimiter = '\t')
N = len(energia)
T = np.linspace(0,5, N)

plt.plot(T, energia, 'r.')
plt.title('Energia vs T para N = %d'%(N))
plt.xlabel('T')
plt.savefig('test_energia.png',format = 'png', dpi = 200)

plt.figure()
plt.plot(T, magnet, 'b.')
plt.title('Magnetizacion vs T para N = %d'%(N))
plt.xlabel('T')
plt.savefig('test_magnet.png',format = 'png', dpi = 200)
plt.show()
```
